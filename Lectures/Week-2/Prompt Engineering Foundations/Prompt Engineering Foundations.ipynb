```python
def calculate_average(numbers):

    if not numbers:

        return None

    total = sum(numbers)

    count = len(numbers)

    return total / count
result = calculate_average([1, 2, 3])

print(result)
```

```python
n = int(input().strip())

factors = []
divisor = 2

while n > 1:
    while n % divisor == 0:
        factors.append(divisor)
        n //= divisor
    divisor += 1

print(factors)
```

```python
intervals = eval(input())

# Sort intervals by start time
intervals.sort(key=lambda x: x[0])

merged = []

for interval in intervals:
    if not merged or merged[-1][1] < interval[0]:
        merged.append(interval)
    else:
        merged[-1][1] = max(merged[-1][1], interval[1])

print(merged)
```

```python
s = input().strip()

stack = []
pairs = {')': '(', ']': '[', '}': '{'}

for ch in s:
    if ch in pairs.values():        # opening bracket
        stack.append(ch)
    else:                           # closing bracket
        if not stack or stack[-1] != pairs[ch]:
            print(False)
            break
        stack.pop()
else:
    print(not stack)
```

```python
arr = eval(input())

n = len(arr)
index = {value: i for i, value in enumerate(arr)}

swaps = 0

for i in range(n):
    correct_value = i + 1
    if arr[i] != correct_value:
        swaps += 1

        # index of the value that should be here
        to_swap_idx = index[correct_value]

        # update index map before swapping
        index[arr[i]] = to_swap_idx
        index[correct_value] = i

        # swap elements
        arr[i], arr[to_swap_idx] = arr[to_swap_idx], arr[i]

print(swaps)
```

```python
nums = list(map(int, input().split()))

max_here = min_here = ans = nums[0]

for x in nums[1:]:
    if x < 0:
        max_here, min_here = min_here, max_here

    max_here = max(x, max_here * x)
    min_here = min(x, min_here * x)

    ans = max(ans, max_here)

print(ans)
```
