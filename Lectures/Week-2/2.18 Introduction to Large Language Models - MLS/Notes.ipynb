## 2. Local Models vs Proprietary Models

### Open-Source / Local Models

Examples discussed:

* Mistral
* LLaMA
* Falcon
* Gemma / MedGemma (domain-specific)

**Characteristics**

* Downloaded and executed on local or private infrastructure
* Can run:

  * On local machines
  * In private cloud
  * In Google Colab (virtual private environment)
* No data sent to third-party servers
* No internet required after download
* Greater control over:

  * Weights
  * Fine-tuning
  * Deployment constraints

**Trade-offs**

* Slower inference unless strong GPUs available
* Requires:

  * GPU configuration
  * Model + tokenizer management
  * Security and vulnerability awareness
* Model freshness is user’s responsibility

---

### Proprietary / Paid Models

Examples:

* OpenAI (ChatGPT, GPT-4o mini)
* Gemini
* Anthropic Claude

**Characteristics**

* API-based access
* Hosted and optimized by vendors
* Faster inference
* Higher quality and more consistent outputs
* Frequent model updates

**Trade-offs**

* Requires API keys
* Cost based on token usage
* Data sent to vendor infrastructure
* Less control over internal model behavior

**Purpose**

**Key Insight**

## Samples on how to use LangChain, Local Model vs Remote Model

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import pipeline

# Load the model using Hugging Face pipeline
hf_pipeline_local = pipeline(
    "text-generation",
    model="tiiuae/Falcon3-1B-Instruct",
    device=0,  # Use GPU (-1 for CPU)
    max_new_tokens = 500,
    return_full_text=False
)

# Create the LangChain LLM using the HuggingFace pipeline
llm_local = HuggingFacePipeline(pipeline=hf_pipeline_local)
```

## To use a remote model we need to use the OpenAI API

```python
#Loading the `config.json` file
import json
import os

# Load the JSON file and extract values
file_name = 'config.json'
with open(file_name, 'r') as file:
    config = json.load(file)
    os.environ['OPENAI_API_KEY'] = config.get("API_KEY") # Loading the API Key
    os.environ["OPENAI_BASE_URL"] = config.get("OPENAI_API_BASE") # Loading the API Base Url
```

## Now we instantiate a remote model from OpenAI

```python
from langchain_openai import ChatOpenAI
llm_remote = ChatOpenAI(
    model="gpt-4o-mini",
    max_tokens=512,
    temperature=0,
    top_p=0.95,
    frequency_penalty=1.2
    )
```

```python
user_input = "How does photosynthesis work?"
response = llm_local.invoke(user_input)
print(response)
```

```python
response = llm_remote.invoke(user_input)
print(response.content)
```

**Above you will see the difference between the answers from a local model and a remote model.**
