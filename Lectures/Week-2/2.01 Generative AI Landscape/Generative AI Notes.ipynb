# Study Notes: What Is Generative AI?

## 1. Artificial Intelligence (AI) — the outer set

**Artificial Intelligence** is the broadest category.
It includes **any system designed to perform tasks that typically require human intelligence**, such as reasoning, learning, perception, decision-making, and language use.

AI is not a single technique. It is an umbrella covering:

* Rule-based systems
* Statistical models
* Learning-based systems
* Hybrid approaches

Everything shown in the slide exists *inside* AI.

---

## 2. Symbolic AI (Rule-Based AI)

**Symbolic AI** represents the earliest form of AI.

### Key characteristics

* Uses **explicit rules**, logic, and symbols
* Knowledge is human-defined
* Reasoning is deterministic and explainable

### Examples

* Expert systems
* Logic programming
* If–then rule engines
* Ontologies and knowledge graphs (without learning)

### Strengths

* High interpretability
* Strong control and determinism
* Good for compliance-heavy domains

### Limitations

* Brittle
* Hard to scale
* Cannot generalize beyond defined rules

Symbolic AI overlaps with AI broadly but is largely **independent of machine learning**.

---

## 3. Machine Learning (ML)

**Machine Learning** is a subset of AI where systems learn patterns from data instead of being explicitly programmed.

### Key idea

> Performance improves with experience (data).

### Core paradigms

* Supervised learning
* Unsupervised learning
* Reinforcement learning

### Outputs

* Predictions
* Classifications
* Rankings
* Scores

### Examples

* Spam detection
* Recommendation systems
* Fraud detection
* Forecasting

Machine Learning overlaps partially with:

* Symbolic AI (hybrid systems)
* Data Science (shared methods and tooling)

---

## 4. Deep Learning (DL)

**Deep Learning** is a subset of Machine Learning based on **multi-layer neural networks**.

### Key characteristics

* Learns hierarchical representations
* Requires large datasets and compute
* Excels at perception and language tasks

### Typical model families

* Neural networks
* Transformers
* CNNs, RNNs (historical)

### Strengths

* Strong generalization
* Handles unstructured data (text, images, audio)
* Enables end-to-end learning

### Limitations

* Low interpretability
* High computational cost
* Sensitive to data quality

Deep Learning sits *inside* Machine Learning and is the **technical foundation of Generative AI**.

---

## 5. Data Science (adjacent but overlapping)

**Data Science** is not AI itself, but it overlaps with AI and ML.

### Focus

* Extracting insights from data
* Statistical analysis
* Visualization
* Experimentation

### Relationship to AI

* Data Science supplies:

  * Clean data
  * Features
  * Evaluation methods
* AI systems depend on data science practices but go beyond analysis into **automation and decision-making**

This explains why Data Science partially overlaps but is not fully contained within AI.

---

## 6. Generative AI (GenAI)

**Generative AI** is a **subset of Deep Learning**, and therefore also a subset of Machine Learning and AI.

### Defining feature

> Generative AI **creates new content**, rather than only predicting or classifying.

### Types of generated content

* Text
* Code
* Images
* Audio
* Video
* Structured data

### How it works

* Learns the **probability distribution of data**
* Generates new samples that resemble training data
* Uses large neural networks (mostly transformer-based)

### Core model types

* Large Language Models (LLMs)
* Vision-language models
* Diffusion models
* Multimodal foundation models

---

## 7. Why Generative AI is shown as a small circle

The slide deliberately shows **Generative AI as a smaller region** because:

* It is **not all of AI**
* It relies heavily on:

  * Deep learning
  * Massive data
  * High compute
* It does not replace:

  * Symbolic reasoning
  * Classical ML
  * Deterministic systems

In real systems, Generative AI is usually **combined** with:

* Rules
* Workflows
* Search
* Databases
* APIs

# Study Notes: Symbolic AI

## 1. What Symbolic AI Is

**Symbolic AI** (also called *classical AI* or *GOFAI – Good Old-Fashioned AI*) is an approach to artificial intelligence based on **explicit symbols and formal logic**.

The core assumption is:

> Intelligence can be achieved by manipulating symbols according to well-defined rules.

Symbolic AI does **not learn from data**. All knowledge and reasoning procedures are **manually designed by humans**.

---

## 2. Core characteristics (as shown on the slide)

### Deterministic rules

* Given the same input, the system always produces the same output
* No randomness or probabilistic behavior
* Behavior is predictable and reproducible

### Formalism

* Uses formal languages such as:

  * Propositional logic
  * First-order logic
  * Description logics
* Reasoning follows strict syntactic and semantic rules

### Mathematical proof

* Conclusions can often be proven correct
* Reasoning steps are traceable and auditable
* Systems can explain *why* a conclusion was reached

---

## 3. Main building blocks (bullet points explained)

### Rules

* Typically expressed as **IF–THEN** statements
  Example:
  `IF patient_has_fever AND patient_has_rash THEN suspect_measles`
* Used heavily in expert systems
* Knowledge is explicit and human-readable

---

### Look-up tables

* Direct mappings from inputs to outputs
* No inference or generalization
* Extremely fast but inflexible
* Common in configuration systems and legacy automation

---

### Search algorithms

* Explore a predefined state space
* Examples:

  * Breadth-first search
  * Depth-first search
  * A*
* Widely used in planning, pathfinding, and game-playing systems
* Intelligence comes from **search + heuristics**, not learning

---

### Ontologies

* Structured representations of knowledge:

  * Concepts
  * Relationships
  * Constraints
* Enable reasoning over domain knowledge
* Common in enterprise knowledge graphs and semantic web systems

---

## 4. Strengths of Symbolic AI

* **Explainability**: Every decision can be traced
* **Control**: Behavior is tightly constrained
* **Reliability**: No hallucinations
* **Compliance-friendly**: Suitable for regulated domains
* **Determinism**: Essential for safety-critical systems

---

## 5. Limitations of Symbolic AI

* **No learning**: Cannot adapt without manual updates
* **Knowledge engineering bottleneck**: Rules are expensive to create and maintain
* **Poor scalability**: Rule interactions become complex
* **Brittleness**: Fails outside predefined scenarios
* **Weak handling of ambiguity and unstructured data**

---

## 6. Relationship to Generative AI (as shown in the diagram)

The slide shows **Generative AI as a small circle partially adjacent to Symbolic AI**, highlighting that:

* Generative AI is **not symbolic by nature**
* GenAI relies on statistical learning, not rules
* However, the two can be **combined in hybrid systems**

### Typical hybrid pattern

* Symbolic AI:

  * Enforces constraints
  * Validates outputs
  * Controls workflows
* Generative AI:

  * Produces text, code, or reasoning steps
  * Handles ambiguity and unstructured inputs

Symbolic AI often acts as the **guardrail layer** around Generative AI.

---

## 7. Symbolic AI vs Generative AI (exam-style contrast)

| Aspect           | Symbolic AI             | Generative AI               |
| ---------------- | ----------------------- | --------------------------- |
| Knowledge        | Explicit, human-defined | Implicit, learned from data |
| Behavior         | Deterministic           | Probabilistic               |
| Explainability   | High                    | Low to medium               |
| Adaptability     | Low                     | High                        |
| Data requirement | Minimal                 | Massive                     |
| Failure mode     | Rigid errors            | Hallucinations              |

---

## 8. Why Symbolic AI still matters

Despite the rise of deep learning and generative models, Symbolic AI remains critical for:

* Business rules
* Policy enforcement
* Safety constraints
* Formal reasoning
* Verification and validation

Modern AI systems increasingly resemble:

> **Symbolic control + Generative reasoning**

---

## 9. One-sentence summary

> Symbolic AI is a deterministic, rule-based approach to intelligence grounded in formal logic and mathematical reasoning, offering explainability and control but limited adaptability, and it increasingly serves as a stabilizing counterpart to generative AI in modern systems.

# Study Notes: Symbolic AI and Language Structure

## 1. Purpose of the slide

This slide uses a **sentence parsing example** to illustrate how Symbolic AI represents and reasons about language using **explicit structure and rules**, rather than statistical learning.

The goal is to show:

* Language as a **formal system**
* Meaning derived from **syntax + roles**
* Intelligence as **symbol manipulation**

---

## 2. The sentence example explained

**Sentence**

> *Dr. McCulloh speaks about industry application of artificial intelligence.*

The sentence is decomposed into **symbolic components**:

### Phrase-level structure

* **Noun phrase**: *Dr. McCulloh*
* **Verb**: *speaks*
* **Object / complement**: *about industry application of artificial intelligence*
* **Prepositional phrase**: *of artificial intelligence*
* **Verb phrase**: *speaks about industry application of artificial intelligence*

This reflects a **grammar-based representation**, not learned embeddings.

---

## 3. How Symbolic AI handles language

### Representation

* Words are symbols
* Phrases are structured objects
* Grammar defines legal compositions

Example (conceptual):

* `NOUN_PHRASE → TITLE + NAME`
* `VERB_PHRASE → VERB + PREPOSITION + NOUN_PHRASE`

Meaning emerges from **structure**, not probability.

---

## 4. Reasoning enabled by symbolic structure

Because the structure is explicit, a symbolic system can:

* Identify **who** performed an action
* Identify **what** action was performed
* Identify **what the action is about**
* Apply logical constraints or transformations

Examples:

* Query: *Who speaks?* → `Dr. McCulloh`
* Query: *What is the topic?* → `industry application of artificial intelligence`
* Rewrite sentence while preserving meaning
* Validate grammatical correctness

---

## 5. Why this mattered historically

Symbolic NLP enabled:

* Early machine translation
* Rule-based chatbots
* Question answering systems
* Formal language understanding

These systems assumed:

> Understanding language = understanding its grammatical structure.

---

## 6. Limitations revealed by the example

Despite its clarity, this approach struggles with:

* Ambiguity
  (“about industry application” could be interpreted in multiple ways)
* Synonyms and paraphrases
  (“talks about” vs “speaks on”)
* Scale
  Grammar and rules explode combinatorially
* Informal language

This is where probabilistic and neural methods outperform symbolic ones.

---

## 7. Hybrid approach highlighted by the slide text

The bullet points emphasize a **modern hybrid view**:

### “The application of rules and formalism is often used together with probabilistic machine learning to improve performance.”

Meaning:

* Rules enforce correctness
* ML handles ambiguity and variation

### “Symbolic AI often provides guardrails for probabilistic approaches.”

Symbolic AI is used to:

* Constrain outputs
* Enforce policies
* Validate structure
* Prevent unsafe or illogical actions

---

## 8. Symbolic AI vs probabilistic language models

| Aspect              | Symbolic NLP       | Probabilistic / GenAI NLP |
| ------------------- | ------------------ | ------------------------- |
| Language model      | Grammar rules      | Statistical patterns      |
| Understanding       | Explicit structure | Implicit representations  |
| Explainability      | High               | Low                       |
| Robustness to noise | Low                | High                      |
| Generalization      | Poor               | Strong                    |

---

## 9. Why this still matters in modern GenAI systems

Even large language models rely on symbolic concepts indirectly:

* Structured outputs (JSON schemas)
* Function calling
* Typed APIs
* Workflow validation
* Policy enforcement

Symbolic reasoning is often **externalized** into control systems rather than embedded in the model.

> This slide demonstrates how Symbolic AI treats language as a formally structured system of symbols and rules, enabling precise reasoning and guardrails that are now commonly combined with probabilistic and generative models in modern AI systems.


## Foundation Models

**Foundation models** are large, general-purpose machine learning models trained on massive, diverse datasets and designed to be **adapted to a wide range of downstream tasks** rather than built for a single use case.

They represent a shift from *task-specific models* to *general representations* that can be reused, composed, and specialized.

---

## 1. Core definition

A foundation model is characterized by:

* **Scale**
  Trained on very large datasets (often multi-domain and multi-modal)

* **Generalization**
  Learns broad representations that transfer across tasks

* **Adaptability**
  Can be specialized via:

  * Prompting
  * Fine-tuning
  * Parameter-efficient tuning (LoRA, adapters)
  * Tool and retrieval augmentation

* **Pretraining → adaptation paradigm**
  One expensive pretraining phase, many cheap downstream adaptations

---

## 2. Why they are called “foundation” models

They act as a **base layer** on which many applications are built:

* Chat systems
* Search and question answering
* Coding assistants
* Image generation
* Agents and workflows
* Enterprise copilots

Instead of training a new model for each task, teams **build on top of the same foundation**.

---

## 3. Model families commonly considered foundation models

### Language

* Large Language Models (LLMs)
* Code models

### Vision

* Vision transformers
* Image diffusion models

### Multimodal

* Text + image
* Text + audio
* Text + video

### Specialized foundations

* Embedding models
* Reranking models
* Speech models

---

## 4. Training characteristics

### Data

* Web-scale text
* Code repositories
* Images, audio, video
* Synthetic data (increasingly important)

### Objectives

* Next-token prediction
* Masked modeling
* Contrastive learning
* Diffusion objectives (for images/video)

### Compute

* Large GPU/TPU clusters

* Long training cycles

## 5. Foundation models vs traditional ML models

| Aspect        | Traditional ML | Foundation Models  |
| ------------- | -------------- | ------------------ |
| Scope         | Narrow task    | Broad, general     |

## 6. Emergent capabilities

At scale, foundation models exhibit behaviors **not explicitly programmed**, such as:

* In-context learning
* Zero-shot and few-shot reasoning
* Cross-domain transfer
* Tool use (when integrated)
* Multimodal grounding

These capabilities emerge from **scale and diversity**, not explicit rules.

## 7. Relationship to Generative AI

Generative AI systems are **almost always built on foundation models**.

## Why transformer models are called **“Transformers”**

The name **Transformer** comes from the **core operation the model performs on data**:
it **transforms sequences into contextual representations using attention**, rather than processing them step by step.

The term was introduced in the 2017 paper *“Attention Is All You Need”*.

---

## 1. What is being “transformed”

Transformers take an **input sequence** (tokens, words, image patches, audio frames) and **transform it into a new sequence of vectors** where:

* Each element encodes **global context**
* Relationships between all elements are explicitly modeled
* Position and meaning are jointly represented

This transformation is:

* Learned
* Parallel
* Context-aware

---

## 2. The role of self-attention

The defining mechanism is **self-attention**.

For each token, the model:

* Compares it with every other token
* Computes relevance scores
* Builds a weighted combination of the entire sequence

This allows each token to:

> “Pay attention” to the parts of the input that matter most for its meaning.

The sequence is thus **transformed from isolated symbols into contextualized representations**.

---

## 3. Why not just “attention models”

Attention existed before transformers (e.g., in RNNs), but:

* Earlier models used attention as a **component**
* Transformers made attention the **primary computation**

The model architecture itself is built around attention layers stacked together, hence a new name was warranted.

---

## 4. What is being replaced

Transformers replaced **recurrent** and **convolutional** sequence models.

| Earlier models | How they process data      |
| -------------- | -------------------------- |
| RNNs / LSTMs   | Sequential, step-by-step   |
| CNNs           | Local, sliding windows     |
| Transformers   | Global, parallel attention |

The “transform” refers to the **change in how sequence information is modeled**.

---

## 5. The mathematical meaning of “transform”

In mathematics and signal processing, a *transform*:

* Maps data from one representation space to another
* Reveals structure not visible in the original space

Transformers do exactly this:

* Input tokens → embedding space
* Embeddings → contextual representation space
* Repeated through stacked layers

Each layer performs a learned transformation.

---

## 6. Encoder–decoder transformation

In the original design:

* The **encoder** transforms input into a latent representation
* The **decoder** transforms that representation into output tokens

This reinforces the idea of **representation transformation**, not just prediction.

---

## 7. Why the name stuck

The term “Transformer” captured several key ideas at once:

* Transformation of representations
* Non-sequential processing
* General-purpose architecture
* Applicability beyond language

The name proved flexible enough to extend to:

* Vision Transformers
* Audio Transformers
* Multimodal Transformers

---

## 8. One-sentence summary

> Transformer models are called “transformers” because they transform entire sequences into context-rich representations using attention-based, parallel, learned mappings rather than step-by-step processing.

# Study Notes: Vector Embeddings

## 1. What a vector embedding is

A **vector embedding** is a numerical representation of an object (text, image, audio, code, user, product, etc.) as a **fixed-length vector of numbers**.

The core idea:

> Similar objects are mapped to vectors that are close to each other in a high-dimensional space.

In the slide:

* *Object 1, 2, 3* → inputs
* *Embedding Model* → transformation
* Numeric vectors → outputs

---

## 2. What counts as an “object”

An object can be:

* A word
* A sentence or paragraph
* A document
* An image
* A piece of code
* An audio clip
* A database record

The embedding model abstracts away raw form and captures **meaning** or **semantic properties**.

---

## 3. What the embedding model does

The embedding model performs a learned transformation:

```
Object → Vector [x₁, x₂, x₃, …, xₙ]
```

### Key properties

* Fixed dimensionality (e.g., 384, 768, 1536)
* Continuous values (floats)
* Learned from large datasets
* Task-agnostic (general-purpose)

Each number has **no standalone human meaning**; meaning exists in the **geometry** of the space.

---

## 4. Why vectors instead of symbols

Symbolic representations:

* Treat items as discrete
* No notion of similarity unless manually defined

Vector embeddings:

* Encode similarity naturally
* Enable distance-based reasoning
* Allow smooth generalization

This enables operations like:

* “Find things like this”
* “Group similar items”
* “Rank by relevance”

---

## 5. Geometry of embedding space

Embedding spaces have useful geometric properties:

### Distance

* Closer vectors → more semantically similar
* Common metrics:

  * Cosine similarity
  * Dot product
  * Euclidean distance

### Clustering

* Similar objects form clusters
* Topics, intents, or concepts emerge naturally

### Directional meaning

* Differences between vectors can encode relationships

---

## 6. How embeddings are trained (conceptually)

Embedding models are trained to:

* Pull related items closer together
* Push unrelated items farther apart

Training signals may come from:

* Co-occurrence in text
* Contrastive learning
* Supervised labels
* Multimodal alignment (text ↔ image)

---

## 7. Why embeddings are not generative

Embeddings:

* **Represent** meaning
* Do not produce text, images, or audio
* Are typically read-only outputs

They are:

* Analytical
* Foundational
* Used by other systems (search, retrieval, ranking)

This is why embedding models are often considered **foundation components**, even though they do not generate content.

---

## 8. Key use cases

### Semantic search

* Query → embedding
* Compare with stored embeddings
* Retrieve closest matches

### Retrieval-Augmented Generation (RAG)

* Retrieve relevant context using embeddings
* Feed results into a generative model

### Recommendation systems

* User embeddings
* Item embeddings
* Match based on similarity

### Clustering and classification

* Group documents
* Detect topics
* Identify anomalies

---

## 9. Why vector embeddings scale well

* Computation is efficient
* Indexing techniques enable fast similarity search
* Same embedding space can serve many applications
* Decouples **understanding** from **generation**

This makes embeddings a core primitive in enterprise AI systems.

---

## 10. Relationship to transformers and foundation models

Modern embedding models are usually:

* Transformer-based
* Derived from foundation models
* Trained once, reused everywhere

They act as:

> A shared semantic coordinate system for AI applications.

---

## 11. One-sentence summary

> Vector embeddings transform objects into fixed-length numerical vectors that encode semantic meaning in geometric space, enabling similarity, retrieval, clustering, and grounding across modern AI systems.

Below are **study notes explaining “zero-shot”, “one-shot”, “few-shot”, and “SOTA”** in the context of **LLM fine-tuning and usage**, with clear distinctions between *prompting* and *training*.

# Zero-shot, One-shot, Few-shot, and SOTA (LLM Context)

## 1. Zero-shot learning

### Definition

**Zero-shot** means the model performs a task **without being given any examples** of that task in the prompt.

* No demonstrations
* Only an instruction or question
* Relies entirely on pretrained knowledge

### Example

> *“Classify the sentiment of this text as positive or negative.”*

The model has never seen examples in the prompt but still attempts the task.

### Why it works in LLMs

* Foundation models learn general patterns during pretraining
* Instruction tuning enables task generalization

### Limitations

* Lower accuracy on complex or domain-specific tasks
* More sensitive to prompt phrasing

---

## 2. One-shot learning

### Definition

**One-shot** means the model is given **exactly one example** of the task in the prompt before being asked to perform it.

### Example

> *Example:
> Text: “I love this product” → Positive
>
> Now classify:
> Text: “This is terrible” → ?*

### Purpose

* Clarifies task format
* Reduces ambiguity
* Improves output consistency

### Trade-off

* Uses more context tokens
* Still limited generalization

---

## 3. Few-shot learning

### Definition

**Few-shot** means the model is given **a small number of examples** (typically 2–10) in the prompt.

### Example

Multiple labeled examples are provided before the query.

### Benefits

* Strong performance gains without training
* Improves formatting and reasoning
* Mimics fine-tuning behavior

### Costs

* Consumes context window
* Performance depends on example quality and order

### Key insight

Few-shot prompting is **not fine-tuning**—it is **in-context learning**.

---

## 4. Prompting vs fine-tuning (important distinction)

| Technique   | Model weights change? | Examples provided |
| ----------- | --------------------- | ----------------- |
| Zero-shot   | No                    | 0                 |
| One-shot    | No                    | 1                 |
| Few-shot    | No                    | Few               |
| Fine-tuning | Yes                   | Thousands+        |

Zero/one/few-shot methods influence **model behavior at inference time only**.

---

## 5. How this relates to fine-tuning

Fine-tuning:

* Updates model weights
* Requires labeled datasets
* Produces consistent behavior without long prompts

Zero/one/few-shot:

* Do not update weights
* Are faster to experiment with
* Are cheaper but less stable

Modern workflows often:

* Prototype with few-shot prompts
* Fine-tune once patterns stabilize

---

## 6. SOTA (State of the Art)

### Definition

**SOTA** stands for **State of the Art**.

It refers to:

* The **best known performance** on a benchmark or task
* At a given point in time

### In LLM fine-tuning context

SOTA may refer to:

* Best zero-shot accuracy
* Best few-shot performance
* Best fine-tuned model on a dataset
* Best efficiency-to-performance trade-off

### Important nuance

SOTA is:

* Benchmark-dependent
* Time-dependent
* Often short-lived

---

## 7. SOTA vs practical usefulness

A model can be SOTA but:

* Too expensive
* Too slow
* Too large
* Too opaque

Production systems often prioritize:

* Reliability
* Cost
* Latency
* Controllability

---

## 8. Common misconceptions

* ❌ Few-shot = fine-tuning
* ❌ Zero-shot means “no training”
* ❌ SOTA means “best for all use cases”

---

## 9. One-sentence summaries

* **Zero-shot**: Task performance with no examples in the prompt
* **One-shot**: Task performance with one example in the prompt
* **Few-shot**: Task performance with a small number of examples in the prompt
* **SOTA**: Best reported performance on a specific task or benchmark at a given time

---

## 10. Exam-ready takeaway

> Zero-shot, one-shot, and few-shot describe how many task examples are provided at inference time to guide an LLM without changing its weights, while SOTA refers to the current best benchmarked performance, often achieved through scale, fine-tuning, or alignment techniques.




# Study Notes: Fine-Tuning Steps for LLMs

## 1. Define the task and objectives

This step clarifies **what problem the model must solve**.

Examples:

* Text generation
* Sentiment analysis
* Question answering
* Classification
* Summarization

Why it matters:

* Determines model architecture choices
* Influences data labeling
* Drives evaluation strategy

A poorly defined task leads to misleading results, even with a strong model.

---

## 2. Set clear performance metrics

Metrics define **what “good” looks like**.

Examples:

* Accuracy → classification
* Precision / Recall / F1 → imbalanced datasets
* BLEU / ROUGE → text generation
* Exact Match → QA

Why it matters:

* Enables objective comparison
* Prevents overfitting to subjective impressions
* Aligns training with business or research goals

---

## 3. Select a pre-trained model

Choose a foundation model that best matches the task.

Examples:

* Encoder-only models → understanding tasks
* Encoder–decoder models → translation, summarization
* Decoder-only models → generative tasks

Why it matters:

* Architecture constrains capability
* Model size affects cost and performance
* Pretraining domain influences results

Fine-tuning adapts capability; it does not create it from scratch.

---

## 4. Prepare the dataset

This is often the **most time-consuming step**.

Includes:

* Cleaning noisy data
* Removing duplicates
* Labeling or verifying labels
* Splitting into:

  * Training
  * Validation
  * Test sets

Why it matters:

* Model quality is bounded by data quality
* Prevents data leakage
* Enables unbiased evaluation

---

## 5. Set up the environment

Prepare the technical infrastructure.

Includes:

* Hardware (GPU / TPU)
* Frameworks (PyTorch, TensorFlow)
* Drivers and dependencies

Why it matters:

* Fine-tuning is compute-intensive
* Environment issues can invalidate experiments
* Reproducibility depends on stable setups

---

## 6. Tokenize the data

Convert raw inputs into model-readable format.

Includes:

* Tokenization
* Padding and truncation
* Attention masks
* Special tokens

Why it matters:

* Tokenization defines how the model “sees” language
* Mismatched tokenizers break performance
* Impacts sequence length and memory usage

---

## 7. Initialize the model

Load pretrained weights and adapt the architecture if needed.

Examples:

* Add classification heads
* Resize output layers
* Freeze or unfreeze layers

Why it matters:

* Controls how much the model can change
* Prevents catastrophic forgetting
* Balances learning speed and stability

---

## 8. Configure training parameters

Set hyperparameters that govern learning.

Common parameters:

* Learning rate
* Batch size
* Number of epochs
* Weight decay

Why it matters:

* Too high → instability
* Too low → slow or stalled learning
* Poor settings waste compute and data

---

## 9. Choose a suitable loss function

Defines **what the model is optimized to minimize**.

Examples:

* Cross-entropy loss → classification
* Token-level loss → language modeling
* Contrastive loss → embeddings

Why it matters:

* Loss function encodes task objectives mathematically
* Incorrect loss leads to wrong optimization behavior

---

## 10. Fine-tune the model

This is the actual training phase.

What happens:

* Forward pass
* Loss computation
* Backpropagation
* Weight updates

Why it matters:

* Model adapts to domain/task
* Risk of overfitting appears here
* Requires monitoring and checkpoints

---

## 11. Evaluate the model

Assess performance on unseen data.

Includes:

* Validation metrics
* Test set evaluation
* Error analysis

Why it matters:

* Confirms generalization
* Identifies failure modes
* Determines readiness for deployment

Evaluation is not optional—it validates the entire process.

---

## Key conceptual takeaway

Fine-tuning is **not a single action**, but a controlled pipeline where:

* Task definition guides design
* Data quality dominates outcomes
* Hyperparameters shape learning
* Evaluation validates success

---

## One-sentence summary

> Fine-tuning an LLM is a structured process that adapts a pretrained foundation model to a specific task through careful task definition, data preparation, parameter configuration, training, and rigorous evaluation.


# Technical Approaches to Fine-Tuning LLMs

## 1. Full fine-tuning (end-to-end fine-tuning)

### What it is

All model parameters are updated using task-specific data.

### Characteristics

* Updates every weight in the model
* Highest flexibility
* Highest compute and memory cost

### Pros

* Maximum task adaptation
* Best possible performance ceiling
* Suitable for highly specialized domains

### Cons

* Very expensive
* Risk of catastrophic forgetting
* Requires large labeled datasets

### Typical use

* Training domain-specific foundation models
* Research or large-scale enterprise customization

---

## 2. Partial fine-tuning (layer-wise fine-tuning)

### What it is

Only a subset of layers is updated, usually the top layers.

### Characteristics

* Early layers frozen
* Later layers adapted

### Pros

* Lower cost than full fine-tuning
* Preserves general language knowledge

### Cons

* Limited adaptation capacity
* Requires careful layer selection

### Typical use

* Moderate domain shifts
* Smaller datasets

---

## 3. Adapter-based fine-tuning

### What it is

Small trainable modules (“adapters”) are inserted between layers, while original weights remain frozen.

### Characteristics

* Original model unchanged
* Adapters store task-specific knowledge

### Pros

* Very parameter-efficient
* Multiple tasks can share one base model
* Easy to swap adapters

### Cons

* Slight performance gap vs full fine-tuning

### Typical use

* Multi-task systems
* Enterprise environments with many variants

---

## 4. LoRA (Low-Rank Adaptation)

### What it is

A parameter-efficient method that injects low-rank matrices into attention layers.

### Characteristics

* Only low-rank matrices are trained
* Base weights are frozen
* Very memory efficient

### Pros

* Excellent performance-to-cost ratio
* Fast training
* Widely supported

### Cons

* Still task-specific
* Less expressive than full fine-tuning

### Typical use

* Customization of large open-weight models
* On-prem or limited-compute setups

---

## 5. Prefix tuning / Prompt tuning

### What it is

Learns a small number of **virtual tokens** prepended to inputs.

### Characteristics

* No model weights updated
* Only prompt vectors are learned

### Pros

* Extremely lightweight
* No risk of model degradation
* Very fast training

### Cons

* Limited adaptation power
* Sensitive to prompt length

### Typical use

* Lightweight personalization
* Rapid experimentation

---

## 6. Instruction tuning

### What it is

Fine-tuning on datasets of **instruction–response pairs**.

### Characteristics

* Improves instruction following
* Often combined with other tuning methods

### Pros

* Makes models more usable
* Improves zero-shot and few-shot performance

### Cons

* Requires high-quality instruction data
* Alignment-focused, not domain-specific

### Typical use

* Assistant models
* General-purpose LLMs

---

## 7. RLHF (Reinforcement Learning from Human Feedback)

### What it is

Uses human preference data to train a reward model and optimize outputs.

### Characteristics

* Multi-stage process
* Human judgments guide optimization

### Pros

* Improves safety and helpfulness
* Aligns model behavior with human values

### Cons

* Expensive and complex
* Hard to scale

### Typical use

* Chat assistants
* Safety-critical deployments

---

## 8. Domain-adaptive pretraining (DAPT)

### What it is

Continued pretraining on large domain-specific corpora before task fine-tuning.

### Characteristics

* Uses unlabeled data
* Precedes supervised fine-tuning

### Pros

* Strong domain adaptation
* Improves downstream performance

### Cons

* High compute cost
* Requires large corpora

### Typical use

* Legal, medical, scientific domains

---

## 9. Comparison summary

| Approach            | Parameters updated | Cost      | Adaptation strength |
| ------------------- | ------------------ | --------- | ------------------- |
| Full fine-tuning    | All                | Very high | Very high           |
| Partial fine-tuning | Some               | High      | Medium              |
| Adapters            | Few                | Low       | Medium              |
| LoRA                | Very few           | Very low  | Medium–high         |
| Prompt tuning       | Minimal            | Very low  | Low                 |
| Instruction tuning  | Varies             | Medium    | Behavioral          |
| RLHF                | Indirect           | Very high | Alignment           |
| DAPT                | All                | High      | Domain-wide         |

---

## 10. Key conceptual takeaway

Fine-tuning is not a single technique but a **spectrum of adaptation strategies** trading off:

* Cost
* Control
* Performance
* Risk

Modern systems increasingly favor **parameter-efficient methods** combined with **retrieval and orchestration**, rather than full model retraining.

---

## One-sentence summary

> Technical fine-tuning approaches for LLMs range from full weight updates to lightweight parameter-efficient methods like LoRA and prompt tuning, each balancing adaptation strength, cost, and operational complexity.

# Study Notes: Image Generation

## 1. Core idea of image generation

Image generation uses **generative models** to create **new visual content** (images or video) that did not previously exist, based on learned patterns from large datasets.

The conceptual similarity to text generation:

* Both learn probability distributions over data
* Both generate outputs step-by-step (or iteratively)
* Both rely on large-scale training

The key difference lies in **data representation and model architecture**.

---

## 2. What is similar to generative text

Both image and text generation:

* Are trained on large datasets
* Use neural networks
* Learn latent representations
* Generate outputs conditioned on prompts or inputs

In both cases, the model learns:

> “What usually comes next given the current context.”

---

## 3. What is different from text generation

### A. Pre-processing

* Text → tokenization
* Images → pixel normalization, resizing, color space conversion

Images are high-dimensional (millions of pixels), which increases complexity.

---

### B. Data

* Text is discrete and sequential
* Images are continuous and spatial
* Images require labeled or paired datasets for conditioning (e.g., text–image pairs)

This changes how similarity and structure are learned.

---

### C. Model architecture

* Text → transformers dominate
* Images → specialized generative architectures are common
* Spatial locality matters more for images

---

## 4. Major image generation model families

### 1. Generative Adversarial Networks (GANs)

**Concept**

* Two networks compete:

  * Generator creates images
  * Discriminator judges real vs fake

**Strengths**

* Sharp, realistic images
* Fast generation after training

**Limitations**

* Training instability
* Mode collapse
* Hard to scale and control

---

### 2. Variational Autoencoders (VAEs)

**Concept**

* Encode images into a latent space
* Decode latent vectors back into images

**Strengths**

* Stable training
* Structured latent space
* Good for representation learning

**Limitations**

* Images often blurrier
* Lower visual fidelity compared to GANs

---

### 3. Diffusion Models

**Concept**

* Gradually add noise to images during training
* Learn to reverse the noise process during generation

**Strengths**

* High-quality, diverse images
* Stable training
* Strong conditioning (e.g., text-to-image)

**Limitations**

* Slower generation
* High compute cost

Diffusion models currently dominate state-of-the-art image generation.

---

## 5. Text-to-image generation

Modern systems combine:

* Text encoders (often transformer-based)
* Image generation models (usually diffusion-based)

This allows:

* Prompt-based image creation
* Style control
* Semantic alignment between text and image

---

## 6. Extension to video generation

Video generation adds:

* Temporal consistency
* Motion modeling
* Much higher computational cost

Conceptually:

* Video = sequence of images + time dependency

Diffusion models are increasingly used for video as well.

---

## 7. Why diffusion models dominate today

Diffusion models offer:

* Better image quality
* Fewer training instabilities
* Flexible conditioning mechanisms
* Strong scalability

This makes them the preferred choice for modern generative image systems.

---

## 8. Relationship to foundation models

Modern image generators are:

* Large-scale
* Pretrained on massive datasets
* Adaptable via conditioning

They function as **visual foundation models**, analogous to LLMs for text.

---

## 9. One-sentence summary

> Image generation uses specialized generative models—most notably diffusion models—to create synthetic images by learning high-dimensional visual distributions, differing from text generation primarily in data representation, preprocessing, and model architecture.





# Study Notes: Competence vs Hallucination in LLMs

## 1. Why this distinction matters

In modern generative AI, **output fluency is not the same as correctness**.
This slide highlights a core tension in LLMs:

> A model can sound confident and articulate while being factually wrong.

Understanding **competence vs hallucination** is essential for:

* Model evaluation
* Trust and reliability
* Enterprise and safety-critical deployments

---

## 2. Competence

**Competence** refers to a model’s ability to produce **correct, relevant, and reliable outputs**.

### Key attributes (from the slide)

* **Accuracy**
  The output is factually correct.

* **Understanding**
  The model captures the intent and meaning of the input.

* **Consistency**
  The model gives stable answers across similar queries.

* **Generalization**
  The model applies knowledge correctly to new or unseen situations.

### Example (from the slide)

* Question: *“What is the capital of France?”*
* Competent answer: **“Paris.”**

This reflects correct retrieval and application of stored knowledge.

---

## 3. Understanding context (competence in conversation)

Competence is not just about isolated facts.

A competent model:

* Stays on topic
* Respects conversational context
* Connects concepts logically

Example:

* In a discussion on climate change, the model correctly discusses:

  * Carbon emissions
  * Renewable energy
  * Global warming
    without drifting into unrelated topics.

This demonstrates **contextual coherence**, not just factual recall.

---

## 4. Hallucination

**Hallucination** occurs when a model generates content that is **not grounded in facts or evidence**, even if it sounds plausible.

### Key attributes (from the slide)

* **Incorrect**
  Factually wrong information.

* **Fabrication**
  Invented details, sources, or explanations.

* **Inconsistency**
  Conflicting answers across similar questions.

* **Unreliable creativity**
  Fluent but untrustworthy outputs.

Hallucinations are a byproduct of probabilistic generation, not intentional deception.

---

## 5. Why hallucinations happen

LLMs:

* Predict the *most likely next token*
* Do not “know” facts in a symbolic sense
* Optimize fluency and likelihood, not truth

When:

* Knowledge is missing
* Prompts are ambiguous
* Context is insufficient
* The model is pushed beyond training data

→ the model may **fill gaps with plausible-sounding text**.

---

## 6. Competence ≠ No hallucinations

A key exam-level insight:

> A model can be competent in many areas and still hallucinate.

Competence is **probabilistic**, not absolute.

This is why:

* Verification
* Grounding
* Retrieval
* Guardrails

are essential in production systems.

---

## 7. Practical mitigation strategies (conceptual)

To increase competence and reduce hallucinations:

* Retrieval-Augmented Generation (RAG)
* Tool calling and external verification
* Constrained outputs (schemas, rules)
* Confidence calibration (“I don’t know” behavior)
* Symbolic or rule-based guardrails

This reflects the hybrid approach emphasized throughout modern AI design.

---

## 8. Exam-style contrast table

| Aspect          | Competence     | Hallucination |
| --------------- | -------------- | ------------- |
| Accuracy        | High           | Low           |
| Grounding       | Evidence-based | Fabricated    |
| Consistency     | Stable         | Variable      |
| Trustworthiness | Reliable       | Unreliable    |
| Fluency         | Often high     | Often high    |

Fluency alone is **not a signal of correctness**.

---

## 9. Key takeaway sentence

> Competence in LLMs reflects accurate, consistent, and context-aware reasoning, while hallucination refers to fluent but incorrect or fabricated outputs arising from probabilistic generation rather than grounded knowledge.




# Study Notes: Fabrication (Hallucination Subtype)

## 1. What “fabrication” means in generative AI

**Fabrication** is a specific form of hallucination in which a model **invents facts, entities, events, or relationships that do not exist**, while presenting them in a confident and coherent manner.

Key property:

* The information is **entirely made up**, not just partially incorrect or outdated.

Fabrication is especially dangerous because the output often appears **plausible, detailed, and authoritative**.

---

## 2. Breakdown of the slide example

### Prompt

> *“Tell me about the contributions of John Doe to the American Revolution.”*

* The prompt assumes that **John Doe** is a real historical figure.
* The model is implicitly encouraged to fill in missing information.

---

### Response

> *“John Doe was a key figure in the American Revolution, known for his strategic prowess during the Battle of Bunker Hill and his diplomatic efforts with France.”*

This response exhibits classic fabrication:

* Uses real historical events (Battle of Bunker Hill)
* Uses real historical themes (diplomacy with France)
* Connects them to a **nonexistent person**

The model constructs a **coherent narrative from unrelated but real concepts**.

---

### Analysis

> *“John Doe is a fabricated figure with no historical records or contributions to the American Revolution.”*

This confirms:

* The model did not retrieve incorrect facts
* It **invented an entity and associated actions**

---

## 3. Why fabrication occurs

Fabrication arises from how LLMs generate text:

* LLMs predict likely continuations, not verified facts
* When asked about an unknown or nonexistent entity:

  * The model prioritizes fluency
  * It avoids saying “I don’t know”
  * It synthesizes plausible content from similar patterns

The model optimizes **linguistic plausibility**, not historical truth.

---

## 4. Fabrication vs simple factual error

| Aspect           | Fabrication              | Factual Error   |
| ---------------- | ------------------------ | --------------- |
| Entity existence | Invented                 | Real            |
| Confidence       | High                     | Often uncertain |
| Plausibility     | Very high                | Variable        |
| Detectability    | Low without verification | Easier          |
| Risk             | High                     | Moderate        |

Fabrication is more harmful because it creates **false knowledge**, not just incorrect details.

---

## 5. Why fabrication is hard to detect

Fabricated outputs:

* Use correct grammar and style
* Reference real events and concepts
* Match expected narrative structures
* Lack internal contradictions

To a non-expert, fabricated content is often indistinguishable from truth.

---

## 6. Common fabrication triggers

* Prompts about obscure or nonexistent entities
* Overly open-ended historical or technical questions
* Missing or insufficient grounding context
* Pressure for detailed answers
* Ambiguous prompts that assume correctness

---

## 7. Fabrication as a subset of hallucination

Hallucination includes:

* Incorrect facts
* Misattribution
* Logical inconsistencies
* Fabrication

Fabrication is the **most severe form**, because it introduces **entirely false entities or events** into the knowledge space.

---

## 8. Mitigation strategies (conceptual)

To reduce fabrication:

* Encourage uncertainty-aware responses
* Require citations or sources
* Use retrieval-based grounding
* Validate named entities
* Constrain generation with structured schemas
* Combine symbolic verification with generative output

These approaches shift the system from **plausibility-first** to **evidence-first** generation.

---

## 9. Key exam takeaway

> Fabrication occurs when a generative model invents entities or facts that do not exist, producing fluent and plausible-sounding outputs that are entirely ungrounded in reality, making it one of the most dangerous forms of hallucination.




# Study Notes: Minimizing Hallucination in Generative AI

## 1. Why hallucination minimization is necessary

Hallucination is an inherent risk in probabilistic, generative models. Since LLMs optimize for **likelihood and fluency**, not truth, minimizing hallucination requires **system-level controls**, not just better prompts.

This slide outlines **complementary mitigation layers** that operate across the AI lifecycle: data, training, inference, and governance.

---

## 2. Train on high-quality data

### Meaning

Hallucination risk is strongly correlated with **training data quality**.

High-quality data characteristics:

* Accurate
* Curated
* Representative
* Up-to-date
* Low noise and duplication

### Why it helps

* Reduces learned false patterns
* Improves grounding in real facts
* Lowers reliance on spurious correlations

Data quality sets the **upper bound** on model reliability.

---

## 3. Fine-tune with specific objectives

### Meaning

Fine-tuning aligns the model toward **task-specific correctness**, not just general fluency.

Examples:

* Domain-specific datasets
* Instruction tuning
* Preference-based optimization

### Why it helps

* Narrows the output distribution
* Reinforces correct behaviors
* Reduces over-generalization

Fine-tuning reduces hallucination **within a defined scope**, not globally.

---

## 4. Prompt engineering

### Meaning

Carefully structured prompts guide the model’s reasoning and response boundaries.

Techniques:

* Explicit instructions (“If unsure, say ‘I don’t know’”)
* Role constraints
* Step-by-step reasoning
* Output format constraints

### Why it helps

* Reduces ambiguity
* Encourages cautious responses
* Limits creative overreach

Prompt engineering influences **inference-time behavior**, not learned knowledge.

---

## 5. Post-processing

### Meaning

Post-processing validates or filters model outputs **after generation**.

Examples:

* Rule-based validation
* Schema checks
* Entity verification
* Confidence thresholds
* Output filtering

### Why it helps

* Catches errors before delivery
* Enforces structural and factual constraints
* Acts as a safety net

Post-processing is essential in production systems.

---

## 6. Human-in-the-loop (HITL)

### Meaning

Human oversight is introduced at critical decision points.

Common patterns:

* Review before publication
* Escalation for low-confidence outputs
* Feedback loops for correction

### Why it helps

* Humans detect subtle fabrication
* Improves trust in high-risk contexts
* Provides training signals for future improvement

HITL trades automation for reliability.

---

## 7. Regular updates and retraining

### Meaning

Models degrade over time as:

* Facts change
* Language evolves
* New knowledge emerges

### Why it helps

* Reduces outdated or stale responses
* Aligns the model with current reality
* Corrects discovered failure modes

Without updates, hallucination risk increases over time.

---

## 8. Symbolic AI with first-order logic

### Meaning

Symbolic systems apply **explicit rules and logical constraints** to model outputs.

Examples:

* Rule engines
* Knowledge graphs
* First-order logic validators
* Policy enforcement layers

### Why it helps

* Enforces consistency
* Prevents logically invalid outputs
* Grounds generation in formal knowledge

This represents a **hybrid AI approach**:

> Generative flexibility + symbolic correctness

---

## 9. Key conceptual insight

No single technique eliminates hallucinations.

Effective mitigation requires:

* **Data-level controls**
* **Training-time alignment**
* **Inference-time constraints**
* **Post-generation validation**
* **Human and symbolic oversight**

Hallucination minimization is a **systems engineering problem**, not a model-only problem.

---

## 10. Exam-ready summary

> Hallucinations in generative AI are minimized through a layered strategy combining high-quality training data, targeted fine-tuning, careful prompting, post-processing validation, human oversight, continuous retraining, and symbolic logic-based constraints.




# Study Notes: Top 10 Vulnerabilities in LLM-Based Systems

These vulnerabilities arise from the **probabilistic nature of LLMs**, their **integration into software systems**, and their **dependence on data, tools, and external components**. They should be understood as **system-level risks**, not just model weaknesses.

---

## 1. Prompt Injection

### Definition

Prompt injection occurs when an attacker manipulates the input prompt to **override system instructions or alter model behavior**.

### How it happens

* Direct injection: User input explicitly changes instructions
* Indirect injection: Malicious instructions embedded in external data (e.g., documents, web pages)

### Why it is dangerous

* Breaks instruction hierarchy
* Can bypass safeguards
* Causes unintended actions or disclosures

Prompt injection is analogous to **SQL injection**, but for natural language interfaces.

---

## 2. Insecure Output Handling

### Definition

Occurs when LLM outputs are **trusted or executed without validation**.

### Examples

* Rendering LLM output directly in a browser (XSS)
* Passing output to backend systems or shells
* Generating executable code or commands

### Risk

* Cross-site scripting (XSS)
* Server-side request forgery (SSRF)
* Privilege escalation
* Remote code execution

LLM outputs must be treated as **untrusted user input**.

---

## 3. Training Data Poisoning

### Definition

Attackers inject malicious or biased data into training datasets.

### How it happens

* Public web crawls
* Open datasets
* User-generated content

### Impact

* Backdoors in model behavior
* Biased or unsafe outputs
* Reduced reliability and trustworthiness

This risk increases with **open and continuously updated datasets**.

---

## 4. Model Denial of Service (DoS)

### Definition

Attackers intentionally overload the model with **resource-intensive requests**.

### Examples

* Extremely long prompts
* Recursive or looping tasks
* High-frequency requests

### Why LLMs are vulnerable

* High compute and memory cost
* Unpredictable inference time
* Shared infrastructure

Results include service degradation, high costs, or downtime.

---

## 5. Supply Chain Vulnerabilities

### Definition

Security risks introduced through **third-party dependencies** in the LLM lifecycle.

### Sources

* Pretrained models
* Open-source libraries
* Plugins and tools
* External APIs

### Impact

* Compromised components
* Hidden backdoors
* Data leakage

LLM systems often depend on **complex, opaque supply chains**.

---

## 6. Sensitive Information Disclosure

### Definition

LLMs unintentionally reveal **confidential or private information**.

### Causes

* Memorization of training data
* Prompt manipulation
* Poor access controls
* Overly permissive system instructions

### Examples

* Personal data
* Credentials
* Proprietary business information

This creates privacy, legal, and regulatory risks.

---

## 7. Insecure Plugin Design

### Definition

Plugins or tools connected to LLMs lack proper security controls.

### Common issues

* Insufficient input validation
* Weak authentication
* Overly broad permissions

### Consequences

* Unauthorized actions
* Data manipulation
* Remote code execution

Plugins effectively extend the model’s **attack surface**.

---

## 8. Excessive Agency

### Definition

LLM-based systems are given **too much autonomy**.

### Examples

* Unrestricted tool access
* Automatic execution of actions
* Lack of approval checkpoints

### Risk

* Unintended side effects
* Irreversible actions
* Cascading failures

Excessive agency turns probabilistic outputs into **operational risk**.

---

## 9. Overreliance

### Definition

Humans or systems rely on LLM outputs **without verification**.

### Consequences

* Acceptance of incorrect information
* Legal and compliance issues
* Poor decision-making

Overreliance amplifies the impact of hallucinations and fabrication.

---

## 10. Model Theft

### Definition

Unauthorized access, copying, or extraction of proprietary models.

### Methods

* API abuse
* Model inversion
* Output-based extraction attacks

### Impact

* Intellectual property loss
* Competitive disadvantage
* Exposure of sensitive capabilities

Model theft is both a **security and business risk**.

---

## Cross-cutting insight

These vulnerabilities share common characteristics:

* LLMs are **non-deterministic**
* Outputs are **not inherently safe**
* Systems are **composed of many interacting parts**
* Trust boundaries are often unclear

Security failures usually occur at the **interfaces**, not inside the model itself.

---

## Exam-ready synthesis

> The top vulnerabilities in LLM systems span prompt manipulation, insecure output handling, poisoned data, resource abuse, supply chain risks, information leakage, unsafe plugins, excessive autonomy, overreliance, and model theft—highlighting that LLM security is a system-level engineering challenge rather than a purely model-level problem.
