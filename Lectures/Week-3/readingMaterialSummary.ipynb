# 📄 **Summary: *Large Language Models as Data Preprocessors***

**Authors:** Haochen Zhang, Yuyang Dong, Chuan Xiao, Masafumi Oyamada
**Published:** 2023 on arXiv (extended in *VLDB Workshops 2024*) ([arXiv][2])

---

## 🧠 **Core Idea**

This study investigates whether **Large Language Models (LLMs)** — like GPT-4 and GPT-4o — can be used not just for generating text, but also for **data preprocessing**, which is the crucial step of cleaning, correcting, and structuring data before it’s used for analysis or machine learning. ([arXiv][1])

Traditional preprocessing requires manual rules or specialized tools, but the authors explore whether LLMs can perform tasks like **error detection, data imputation, schema matching, and entity matching** through clever prompting and existing language understanding. ([arXiv][1])

---

## 🔍 **Motivation**

Data preprocessing is critical in data mining and analytics but often:

* requires manual rules,
* is expensive to build and maintain,
* depends on domain expertise.

Because LLMs understand natural language and can reason about text data, the authors explore their **capability to automate and improve preprocessing** tasks. ([arXiv][1])

---

## 🧩 **Preprocessing Tasks Addressed**

The paper focuses on four major types:

1. **Error Detection (ED)** – spotting incorrect or inconsistent data entries.
2. **Data Imputation (DI)** – filling in missing values.
3. **Schema Matching (SM)** – aligning different table formats.
4. **Entity Matching (EM)** – identifying whether two records represent the same real-world entity. ([arXiv][1])

---

## 🛠 **Proposed Framework**

Rather than using LLMs out-of-the-box, the authors create a **prompting methodology** that includes:

### **Prompt Engineering Techniques**

* **Zero-shot prompting:** Direct instructions to the model without examples.
* **Few-shot prompting:** Giving the model sample examples to condition it.
* **Batch prompting:** Passing multiple records at once to reduce cost.
* **Contextualization & Feature Selection:** Providing additional context and relevant fields to improve model precision. ([arXiv][1])

These components help the LLM understand the task and produce structured outputs instead of just free text. ([arXiv][1])

---

## 📊 **Experimental Evaluation**

The authors tested this framework on **12 public datasets** across the four preprocessing tasks. Three LLMs were compared:

* **GPT-3.5**
* **GPT-4**
* **GPT-4o**

**Key Findings:** ([arXiv][1])

✅ **GPT-4** generally outperformed GPT-3 and GPT-4o.

* Achieved **100% accuracy or F1 score** in 4 out of 12 datasets.
* Strong performance in *data imputation* and *entity matching*.
* Less consistent on error detection and schema matching.

✅ **GPT-3.5** still performed well and is more cost-effective than GPT-4.

⚠️ **GPT-4o** showed inconsistent results — strong on some tasks but weak on others. ([arXiv][1])

---

## 📈 **Strengths Identified**

The authors highlight several advantages of using LLMs in preprocessing: ([arXiv][1])

* **Semantic understanding:** Can grasp text meaning beyond simple rule matching.
* **No need for hand-engineered rules:** Reduces developer burden.
* **Reasoning ability:** Can provide explanations along with outputs.
* **Adaptable via prompts and examples:** Easier customization for different tasks.

---

## ⚠️ **Limitations & Challenges**

Despite promising results, LLM-based preprocessing has shortcomings: ([arXiv][1])

* **Computational cost:** Running large models on big data is expensive.
* **Inconsistency:** Some tasks (especially schema matching) still see mixed performance.
* **Model hallucination:** LLMs sometimes generate plausible but incorrect outputs.
* **Domain specificity:** Harder to apply in specialized domains without careful prompt design. ([arXiv][1])

---

## 📌 **Conclusions**

The study concludes that LLMs show **significant promise** as data preprocessors, especially for tasks requiring semantic understanding and flexible reasoning. However, the framework needs optimization, and more research is needed to improve efficiency and robustness. ([arXiv][1])

---

## 🧠 **Implication**

This work suggests a new use case for LLMs: **automating data cleaning and preparation**, which could transform workflows in data science and analytics. ([arXiv][2])

# 📄 Tree of Thoughts: Deliberate Problem Solving with Large Language Models

**Authors:** Shunyu Yao, Dian Yu, Jeffrey Zhao, Izhak Shafran, Thomas L. Griffiths, Yuan Cao, Karthik Narasimhan ([arXiv][1])
**Published on:** arXiv, May 17, 2023 (NeurIPS 2023) ([arXiv][1])

---

## 🧠 What the Paper Is About

Large Language Models (LLMs) like GPT-4 are powerful — but when solving **complex reasoning tasks**, the standard inference method (predicting tokens left-to-right) can be insufficient. The authors propose a new **inference framework** called **Tree of Thoughts (ToT)** that allows models to:

👉 Explore **multiple reasoning paths**
👉 Evaluate intermediate decisions
👉 Backtrack and revise choices
👉 Incorporate lookahead planning rather than a single linear thought sequence

This approach helps the model make more **deliberate and strategic decisions**, akin to how humans think when solving puzzles or planning. ([arXiv][1])

---

## 🌳 Core Concept: *Thoughts* and *Search*

### What Is a *Thought*?

A **thought** is a coherent chunk of reasoning — bigger than a token but smaller than the full answer. It serves as an **intermediate decision** or step toward solving a problem. 

### Why a Tree?

Instead of following a single chain of reasoning (like Chain-of-Thought), ToT treats the reasoning process as a **tree search** over thoughts:

* Each node = one thought
* Each branch = a possible continuation
* The model explores, evaluates, and prunes paths
* Allows backtracking when a path seems unpromising 

---

## 🔍 How Tree of Thoughts Works

ToT consists of four key elements:

### 1. **Thought Decomposition**

Choosing how to divide the problem into meaningful intermediate steps.
Example: in a math puzzle, thoughts could be partial equations. 

### 2. **Thought Generation**

At any intermediate state, the model proposes multiple possible next thoughts (branching factor).
Methods:

* Sampling multiple candidates
* Explicitly prompting for candidate steps 

### 3. **Thought Evaluation**

Each candidate is evaluated to estimate how promising it is toward solving the overall problem.
This evaluation itself is done by the LLM, which scores or compares the different partial solutions. 

### 4. **Search Algorithm**

Once thoughts are generated and evaluated, a search algorithm directs traversal of the tree:

* **Breadth-First Search (BFS):** Keeps multiple parallel options
* **Depth-First Search (DFS):** Explores deep paths first
* Optional pruning based on evaluation scores 

This turns LLM reasoning into something similar to classical AI tree search, but driven by the model’s own language-based reasoning rather than external heuristics.

---

## ✅ Why This Matters

Traditional LLM inference methods have limitations:

| Method           | Handles Exploration | Backtracking | Strategic Planning |              |
| ---------------- | ------------------- | ------------ | ------------------ | ------------ |
| Direct Prompting | ❌                   | ❌            | ❌                  |              |
| Chain of Thought | ❌                   | ❌            | ❌                  |              |
| ToT              | ✅                   | ✅            | ✅                  | ([arXiv][1]) |

ToT allows the model to reason like a **planner or search agent**, not just an autoregressive predictor.

---

## 🧪 Tasks Used for Evaluation

The authors test ToT on three challenging benchmarks:

### 1. **Game of 24**

Given 4 numbers, combine them using arithmetic to make 24.

* Chain-of-Thought GPT-4 solved ~4%
* ToT GPT-4 solved ~74% ([arXiv][1])

### 2. **Creative Writing**

Generate coherent multi-paragraph text.
ToT explores multiple writing outlines before selecting the best continuation. ([arXiv][1])

### 3. **Mini Crosswords**

Fill crossword grids requiring vocabulary and logic.
ToT explores multiple possible placements and backtracks when needed. ([arXiv][1])

---

## 🎯 Major Findings

* **ToT dramatically boosts LLM performance** on tasks requiring planning and search. ([arXiv][1])
* It enables **strategic lookahead** and **revision**, unlike standard inference. ([arXiv][1])
* The core idea is general: the same framework works across different types of problems. ([arXiv][1])

---

## 🧩 Why Traditional LLM Reasoning Falls Short

Standard LLM inference predicts one token at a time, which is:

* **Local:** Only cares about the next token
* **Linear:** No branching or second thoughts
* **Greedy:** Early errors cannot be revised ([arXiv][1])

ToT turns reasoning into a **deliberate search process** — like classic problem-solving in AI.

---

## 🧠 Relation to Human Cognition

The authors connect ToT to cognitive theories where humans use:

* **System 1:** Fast, intuitive thinking
* **System 2:** Slow, deliberate planning

ToT aims to simulate “System 2” reasoning with LLMs. 

---

## 🚀 Implications

* ToT represents a **paradigm shift in LLM inference** — from left-to-right generation to structured search. ([arXiv][1])
* It bridges LLMs with classical AI search and planning. ([arXiv][1])
* Opens paths toward **more autonomous agents** and complex problem solvers. ([arXiv][1])

---

## 🔎 Summary

**Tree of Thoughts (ToT)** is a powerful new framework that:

* Explores multiple reasoning paths
* Allows backtracking and lookahead
* Uses LLMs as both generators and evaluators in a search process

This leads to **dramatically better performance** on tasks requiring planning and complex reasoning than traditional prompting strategies. ([arXiv][1])
