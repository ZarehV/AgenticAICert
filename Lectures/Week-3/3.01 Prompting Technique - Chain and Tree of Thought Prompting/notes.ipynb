# Chain of Thought and Tree of Thought Prompting

## 🧠 Chain of Thought (CoT) Prompting

![Image](https://www.promptingguide.ai/_next/image?q=75\&url=%2F_next%2Fstatic%2Fmedia%2FTOT.3b13bc5e.png\&w=3840)

![Image](https://media.licdn.com/dms/image/v2/D4D12AQH-ZCS2JiL5Bg/article-cover_image-shrink_720_1280/B4DZbZdLWuGwAM-/0/1747405027634?e=2147483647\&t=YieeRrO9PdewCvKNsAvSoFCe6vkr-R0AhAYIb3YTSTg\&v=beta)

![Image](https://miro.medium.com/v2/resize%3Afit%3A1400/1%2AFalx2RXFxc3X8U3nTBzkIw.png)

### What it is

**Chain of Thought prompting** encourages the model to **reason step by step in a linear sequence**, instead of jumping straight to an answer.

Instead of:

> *“What’s the answer?”*

You nudge the model with:

> *“Let’s think step by step.”*

This causes the model to generate intermediate reasoning before the final output.

---

### Why it works

Large Language Models are trained on examples where reasoning is written out (math solutions, explanations, proofs). By prompting for reasoning explicitly, you activate that learned pattern.

---

### Example

**Without CoT**

```
Q: If a train travels 60 km in 1 hour, how long to travel 180 km?
A: 3 hours
```

**With CoT**

```
Q: If a train travels 60 km in 1 hour, how long to travel 180 km?
Let's think step by step.
A: 
- The train travels 60 km per hour.
- 180 km ÷ 60 km/hour = 3 hours.
So the answer is 3 hours.
```

---

### Best use cases

* Math and logic problems
* Multi-step reasoning
* Word problems
* Explanations and teaching
* Debugging reasoning errors

---

### Best practices

* Use phrases like:

  * *“Think step by step”*
  * *“Explain your reasoning”*
* Works best with **few-shot examples** that already include reasoning
* Be careful in production apps: exposing reasoning may be undesirable (privacy, safety, verbosity)

---

## 🌳 Tree of Thought (ToT) Prompting

![Image](https://www.promptingguide.ai/_next/image?q=75\&url=%2F_next%2Fstatic%2Fmedia%2FTOT.3b13bc5e.png\&w=3840)

![Image](https://substack-post-media.s3.amazonaws.com/public/images/0f530fe7-d3dc-4bab-ac85-47246323b794_2416x1360.png)

![Image](https://miro.medium.com/v2/resize%3Afit%3A1400/1%2AZaK1ivfNLwAMIbtS5dOUkw.png)

### What it is

**Tree of Thought prompting** extends Chain of Thought by allowing the model to **branch into multiple reasoning paths**, explore alternatives, evaluate them, and then choose the best one.

Think:

* CoT → *one straight line of reasoning*
* ToT → *many branches, prune the bad ones*

---

### Core idea

Instead of:

> *“Reason step by step”*

You ask:

> *“Generate multiple possible approaches, evaluate them, and pick the best.”*

---

### Example (simplified)

```
Problem: How can a company reduce cloud costs?

Approach 1:
- Reduce instance sizes
- Risk: performance degradation

Approach 2:
- Use spot instances
- Risk: interruptions

Approach 3:
- Improve caching and architecture
- Risk: engineering effort

Evaluation:
- Approach 3 provides long-term savings with controlled risk

Final Answer:
- Optimize architecture and caching
```

---

### Why it’s powerful

* Encourages **exploration before commitment**
* Reduces hallucinations by self-evaluation
* Mimics human problem solving (brainstorm → evaluate → decide)

---

### Best use cases

* Strategic planning
* Design decisions
* Complex open-ended problems
* Agentic AI systems
* Research and ideation
* Multi-objective optimization

---

### Best practices

* Explicitly ask for:

  * Multiple options
  * Evaluation criteria
  * Selection rationale
* Often implemented programmatically with:

  * Iterative prompting
  * Scoring or voting
  * Agent loops (e.g., planner → evaluator → decider)

---

## 🔍 Chain of Thought vs Tree of Thought

| Aspect              | Chain of Thought       | Tree of Thought                 |
| ------------------- | ---------------------- | ------------------------------- |
| Reasoning structure | Linear                 | Branching                       |
| Exploration         | Single path            | Multiple paths                  |
| Complexity          | Medium                 | High                            |
| Cost (tokens)       | Lower                  | Higher                          |
| Best for            | Deterministic problems | Ambiguous or strategic problems |
| Common in agents    | Sometimes              | Very common                     |

---

## 🧩 How they fit into Agentic AI

* **CoT** → Internal reasoning inside a single agent step
* **ToT** → Planning, reflection, and decision-making across steps
* ToT often pairs with:

  * Reflection
  * Self-critique
  * Planning + execution loops

---

### One-liner summary

* **Chain of Thought**: *“Show your work.”*
* **Tree of Thought**: *“Try multiple ways, judge them, then decide.”*



````python
from langchain_huggingface import HuggingFacePipeline
from transformers import pipeline

# Load the model using Hugging Face pipeline
hf_pipeline_local = pipeline(
    "text-generation",
    model="tiiuae/Falcon3-1B-Instruct",
    device=0,  # Use GPU (-1 for CPU)
    max_new_tokens = 500,
    return_full_text=False
)

# Create the LangChain LLM using the HuggingFace pipeline
llm_local = HuggingFacePipeline(pipeline=hf_pipeline_local)
````

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough

cot_template_1 = '''Stance classification is the task of determining the expressed or implied opinion, or stance, of the reply towards the comment.
comment: {comment}
reply: {reply}
explanation:'''

cot_prompt_1 = PromptTemplate(
    input_variables=["comment","reply"],
    template=cot_template_1
)

cot_chain_1 = cot_prompt_1 | llm_local

cot_template_2 ='''Therefore, based on your explanation, {stance_reason}, what is the final stance? Respond with a single word: "agree", "disagree", or "neutral". Only return the stance as a single word, and no other text.
comment: {comment}
reply: {reply}
stance:'''

cot_prompt_2 = PromptTemplate(
    input_variables=["comment","reply","stance_reason"],
    template=cot_template_2
)

cot_chain_2 = cot_prompt_2 | llm_local

cot_chain = {"stance_reason": cot_chain_1, "comment":RunnablePassthrough(), "reply":RunnablePassthrough()} | cot_chain_2
````

````python
test_comment = "The new Dune movie does not really capture the vision laid out by Frank Herbert. It feels like they tried to import too many visual effects that take away from the philosophy of the work."
test_replies = [
    "The newer ones fail to live up to the sophistry of the older movies from the 70's.",
    "Frank Herbert wrote a lot of books.",
    "I think the new Dune movie better captures the spirit, if not the content, of Frank Herbert's philosophy.",
    "The quick red fox jumped over the lazy brown dog.",
    "Yeah, this new movie is a real masterpiece, lol!!"
]
````

````python
responses = []
for reply in test_replies:
    response = cot_chain.invoke({"comment": test_comment, "reply": reply})
    responses.append(response)

# Print the results
for idx, (reply, response) in enumerate(zip(test_replies, responses)):
    print(f"Reply {idx+1}: {reply}\nStance: {response}\n")
````

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough

# Step 1: Ask for step-by-step reasoning
cot_template_reason = """Solve the following math problem step by step:
Question: {question}
Reasoning:"""

cot_prompt_reason = PromptTemplate(
    input_variables=["question"],
    template=cot_template_reason
)

# Step 2: Ask for the final answer, using the reasoning
cot_template_answer = """Based on your reasoning: {reasoning}
What is the final answer? Only return the answer as a number.
Question: {question}
Answer:"""

cot_prompt_answer = PromptTemplate(
    input_variables=["question", "reasoning"],
    template=cot_template_answer
)

# Chains
cot_chain_reason = cot_prompt_reason | llm_local
cot_chain_answer = cot_prompt_answer | llm_local

# Compose the chain using RunnablePassthrough to pass the question unchanged
cot_chain_math = {
    "reasoning": cot_chain_reason,
    "question": RunnablePassthrough()
} | cot_chain_answer

# Example usage
math_question = "If a train travels 60 km in 1 hour, how long to travel 180 km?"
result = cot_chain_math.invoke({"question": math_question})
print(result)
````

# Tree-of-Thought Prompt

## 🌳 Tree-of-Thought (ToT) Prompting

![Image](https://www.promptingguide.ai/_next/image?q=75\&url=%2F_next%2Fstatic%2Fmedia%2FTOT.3b13bc5e.png\&w=3840)

![Image](https://learnprompting.org/docs/assets/advanced/advanced_covers/tot_cover.svg)

![Image](https://api.wandb.ai/files/sauravmaheshkar/images/projects/39501000/f1beb662.png)

### Definition

**Tree-of-Thought prompting** is a prompting technique where a Large Language Model is guided to **generate multiple reasoning paths (branches)** for a problem, **evaluate or score those paths**, and then **select the best solution** instead of committing to a single linear chain of reasoning.

In short, the model is asked to:

1. Explore different possible thoughts
2. Assess their quality or feasibility
3. Choose the most promising one

---

### Core intuition

Humans rarely solve hard problems in one straight line—we **branch, backtrack, and compare options**.
Tree-of-Thought prompting explicitly encourages this behavior in LLMs.

* **Chain-of-Thought** → one reasoning path
* **Tree-of-Thought** → many paths + selection

---

### Simple prompt pattern

```
1. Generate multiple possible approaches to solve the problem.
2. Evaluate each approach based on given criteria.
3. Select the best approach.
4. Produce the final answer using that approach.
```

---

### Example

**Prompt**

```
Problem: How should a startup choose a cloud provider?

Generate three possible strategies.
Evaluate pros and cons of each.
Select the best option and explain why.
```

**Model behavior**

* Branches into AWS / GCP / Azure strategies
* Compares cost, ecosystem, scalability
* Chooses the best-fit provider based on constraints

---

### Why Tree-of-Thought works

* Encourages **exploration before commitment**
* Reduces premature conclusions
* Improves reasoning on **ambiguous or strategic tasks**
* Enables **self-evaluation and reflection**
* Aligns naturally with **agentic workflows**

---

### When to use Tree-of-Thought

* Open-ended problems
* Planning and decision-making
* Design and architecture choices
* Research, brainstorming, and ideation
* Multi-objective trade-offs
* Agent-based systems (planner → evaluator → executor)

---

### Implementation styles

**Prompt-only**

* Single prompt asking for branching + evaluation

**Programmatic (Agentic)**

* Multiple LLM calls
* Thought generation → scoring → pruning
* Often combined with:

  * Reflection
  * Voting
  * Heuristics or reward models

---

### Trade-offs

| Advantage           | Cost               |
| ------------------- | ------------------ |
| Better reasoning    | More tokens        |
| Lower hallucination | Higher latency     |
| Stronger decisions  | More orchestration |

---

### One-sentence summary

**Tree-of-Thought prompting asks the model to explore multiple reasoning paths, evaluate them, and choose the best—mirroring how humans solve complex problems.**



````python
# Step 1: Generate Hypotheses
hypothesis_template = '''
Consider the following comment and reply:
comment: {comment}
reply: {reply}

Generate three different hypotheses for the stance of the reply towards the comment.
For each hypothesis, explain why the reply might:
1. Agree
2. Disagree
3. Be Neutral

Output each hypothesis clearly labeled (e.g., "Hypothesis 1: ...") with a newline between each hypothesis.'''

hypothesis_prompt = PromptTemplate(
    input_variables=["comment", "reply"],
    template=hypothesis_template
)

hypothesis_chain = hypothesis_prompt | llm_local

# Step 2: Evaluate Hypotheses
evaluation_template = '''
Consider the following comment and reply:
comment: {comment}
reply: {reply}

Given the following hypotheses and explanations for the stance of the reply towards the comment:
{hypotheses}

Evaluate each hypothesis based on its logical consistency and support from the reply.
Assign a numerical score from 1 to 5 for each hypothesis, where 5 is highly consistent and 1 is inconsistent. Only reply with the score and reason for that score for each hypothesis (e.g., Hypothesis 1: [score], reason: ...) and no other text.'''

evaluation_prompt = PromptTemplate(
    input_variables=["hypotheses", "comment", "reply"],
    template=evaluation_template
)

evaluation_chain = evaluation_prompt | llm_local

# Step 3: Final Decision
decision_template = '''
Consider the following comment and reply:
comment: {comment}
reply: {reply}

Based on the evaluations of different hypotheses for the stance of the reply towards the comment:
{hypotheses}

{evaluations}

Select the hypothesis with the highest score. Output the final stance as "agree", "disagree", or "neutral" based on the chosen hypothesis. Only output the label as single word and do not generate any other text after the label.
label:'''

decision_prompt = PromptTemplate(
    input_variables=["hypotheses", "evaluations", "comment", "reply"],
    template=decision_template
)

decision_chain = decision_prompt | llm_local

# Combine the chains into a Tree-of-Thought pipeline
tot_chain = {
    "hypotheses": hypothesis_chain,
    "comment": RunnablePassthrough(),
    "reply": RunnablePassthrough()
} | RunnablePassthrough.assign(evaluations = evaluation_chain) | decision_chain
````

````python
hypotheses = hypothesis_chain.invoke(
    {
        "comment": test_comment,
        "reply": reply
    }
)
print(hypotheses)
````

````python
evaluations = evaluation_chain.invoke(
    { 
        "hypotheses": hypotheses,
        "comment": test_comment,
        "reply": reply
    }
)
print(evaluations)
````

````python
responses = []
for reply in test_replies:
    response = tot_chain.invoke({"comment": test_comment, "reply": reply})
    responses.append(response)

# Print results
for idx, (reply, response) in enumerate(zip(test_replies, responses)):
    print(f"Reply {idx+1}: {reply}\nStance: {response}\n")
````

## 🌳 Tree of Thought (ToT) Prompting Example

This example demonstrates the Tree of Thought approach:
- Generate multiple solution options for a problem
- Score each option
- Select the best option

You will see the outputs at each step.

````python
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough

# Step 1: Generate multiple options
options_prompt = PromptTemplate(
    input_variables=["problem"],
    template="""Generate three different solution options for the following problem. Number each option.
Problem: {problem}
Options:"""
)

generate_options = options_prompt | llm_local

# Step 2: Score each option
score_prompt = PromptTemplate(
    input_variables=["problem", "option"],
    template="""Given the problem and the option, score the option from 1 (worst) to 10 (best) and briefly justify the score.
Problem: {problem}
Option: {option}
Score and Justification:"""
)

score_option = score_prompt | llm_local

# Step 3: Select the best option
select_prompt = PromptTemplate(
    input_variables=["problem", "scored_options"],
    template="""Given the problem and the scored options below, select the best option and explain why.
Problem: {problem}
Scored Options:
{scored_options}
Best Option and Reasoning:"""
)

select_best = select_prompt | llm_local
````

````python
# Example problem
tot_problem = "How can a company reduce cloud computing costs?"

# Step 1: Generate options
options_output = generate_options.invoke({"problem": tot_problem})
print("Generated Options:\n", options_output)
````

````python
# Parse options (assuming the model returns numbered options)
import re
options = re.findall(r"\d+\. (.*)", options_output)
if not options:
    options = [opt.strip() for opt in options_output.split("\n") if opt.strip()]

# Step 2: Score each option
scored_options = []
for opt in options:
    score = score_option.invoke({"problem": tot_problem, "option": opt})
    scored_options.append(f"Option: {opt}\n{score}")
    print(f"\nOption: {opt}\n{score}")
````

````python
scored_options
````

````python
# Step 3: Select the best option
scored_options_text = "\n\n".join(scored_options)
best = select_best.invoke({"problem": tot_problem, "scored_options": scored_options_text})
print("\nBest Option and Reasoning:\n", best)
````
