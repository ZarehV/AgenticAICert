# 📌 Practice Scope Overview

The firm handles:

1. Divorce (contested & uncontested)
2. Child custody & visitation
3. Child support
4. Spousal support (alimony)
5. Enforcement & modification actions
6. Paternity
7. Prenuptial & postnuptial agreements
8. Protective orders (family violence)
9. Adoption (limited scope)
10. Grandparents’ rights (limited cases)

The firm does NOT handle:

* Criminal defense (except protective-order-related)
* Immigration
* Bankruptcy
* CPS defense (Child Protective Services)
* Appeals
* Out-of-state cases
* Estate planning
* Same-day emergency filings after business hours

# 📂 Detailed Case Scenarios (For Qualification Logic)

Below are categorized scenarios that your Intake + Case Qualification Agent should recognize and classify.

---

## 1️⃣ Divorce Cases

### ✅ Within Scope

* Uncontested divorce (no dispute over property or custody)
* Contested divorce
* High-asset divorce (business ownership, retirement accounts)
* Divorce with minor children
* Divorce involving real estate division
* Military divorce (if state-based)
* Divorce involving domestic violence

### ⚠️ Conditional Qualification

* Divorce where the other spouse already has an aggressive attorney
* Divorce involving hidden assets
* Divorce where one party lives out of state
* Divorce filed in another county within state

### ❌ Out of Scope

* Divorce filed in another state
* Appeals of finalized divorce decrees
* Divorce involving ongoing criminal charges unrelated to family violence

---

## 2️⃣ Child Custody / Conservatorship

### ✅ Within Scope

* Initial custody determination
* Modification of custody
* Emergency custody request
* Visitation enforcement
* Relocation disputes
* Joint managing conservatorship
* Sole managing conservatorship

### ⚠️ Conditional

* Interstate custody disputes (UCCJEA issues)
* Allegations of abuse
* Parent with substance abuse issues

### ❌ Out of Scope

* CPS removal cases
* Dependency court defense

---

## 3️⃣ Child Support

### ✅ Within Scope

* Establish child support
* Modify child support
* Enforce unpaid child support
* Calculate support obligations

### ❌ Out of Scope

* Federal child support enforcement appeals
* Cases outside state jurisdiction

---

## 4️⃣ Spousal Support / Alimony

### ✅ Within Scope

* Temporary spousal support
* Post-divorce alimony
* Modification of support
* Enforcement of alimony

---

## 5️⃣ Protective Orders (Family Violence)

### ✅ Within Scope

* Filing for protective order
* Defense against protective order
* Emergency temporary orders

### ⚠️ Conditional

* Protective order tied to criminal charges

### ❌ Out of Scope

* Criminal defense for assault charges

---

## 6️⃣ Paternity

### ✅ Within Scope

* Establish paternity
* Challenge paternity
* Add father to birth certificate
* Custody/support tied to paternity

---

## 7️⃣ Prenuptial / Postnuptial Agreements

### ✅ Within Scope

* Drafting agreements
* Reviewing agreements
* Enforcement disputes

---

## 8️⃣ Adoption (Limited Scope)

### ✅ Within Scope

* Stepparent adoption
* Private adoption (no CPS involvement)

### ❌ Out of Scope

* Foster system adoption
* CPS adoption cases
* International adoption

---

## 9️⃣ Grandparents’ Rights

### ✅ Within Scope (Selective)

* Grandparent visitation under statutory grounds

### ⚠️ Requires Attorney Review

* Cases without clear statutory basis

# 🧠 Qualification Factors (Important for Agent Design)

Your AI intake agent should evaluate:

### 1. Jurisdiction

* What state?
* What county?
* Where do children reside?
* Where was marriage registered?

### 2. Urgency

* Emergency custody?
* Domestic violence?
* Court date already scheduled?

### 3. Conflict Check

* Opposing party name
* Prior consultations

### 4. Case Complexity Scoring

You can design a scoring model:

* High conflict
* High asset
* Domestic violence
* Substance abuse
* Interstate issues
* Business ownership

### 5. Financial Qualification

* Ability to pay retainer
* Income level
* Willingness to finance
