## 1) User experience and entry points

**FR-1.1 Multi-channel intake**

* Support web chat initially (MVP), with hooks for SMS and voice later.
* Seamless handoff to human staff (live chat / “request callback”) with transcript + summary.

**FR-1.2 User identity + session control**

* Anonymous pre-intake allowed.
* Optional “resume later” via email/SMS magic link.
* Consent gates (disclaimers) before collecting sensitive details.

**FR-1.3 Language + accessibility**

* At minimum English + Spanish support (or configurable languages).
* Reading level simplification mode (“explain in plain English”).

---

## 2) Compliance, consent, and ethics guardrails

**FR-2.1 Legal disclaimers + consent capture**

* Present “not legal advice / no attorney-client relationship” notice.
* Capture explicit user acknowledgement (timestamped).
* Capture consent to store/process data (privacy notice).

**FR-2.2 Emergency/safety routing**

* If user indicates domestic violence, child abuse risk, suicidal ideation, or immediate danger:

  * Provide emergency resources (911 / local hotlines) and stop legal qualification flow.
  * Option to escalate to staff.

**FR-2.3 Unauthorized practice of law safeguards**

* Agent must avoid giving prescriptive legal advice.
* Allowed outputs: procedural info, next steps, and “attorney review required” language.

---

## 3) Structured intake interview (dynamic question engine)

**FR-3.1 Matter-type detection (family law taxonomy)**

* Classify intent early into: divorce, custody/visitation, child support, spousal support, paternity, modification, enforcement, protection order, adoption, prenuptial/postnuptial, property division, relocation, contempt, mediation.
* If unclear: run a clarification branch (2–4 targeted questions).

**FR-3.2 Core fact collection (always-on)**
Collect and validate:

* Names (client + other party), contact info, preferred contact method.
* County + state + where parties/children reside.
* Relationship status (married/common-law/never married), marriage date, separation date.
* Children: names/ages, residence, school district, existing orders.
* Safety: restraining orders, threats, weapons, abuse history (with careful wording).

**FR-3.3 Matter-specific modules (conditional)**
Examples:

* **Divorce**: assets/debts overview, business ownership, real estate, retirement, prenup, date of marriage/separation, jurisdiction.
* **Custody**: current parenting schedule, primary caretaker, relocation intent, substance issues, school/medical decision conflicts.
* **Support**: income info, employment, childcare costs, health insurance, arrears, prior support orders.
* **Modification/Enforcement**: existing order details, what changed, violations, last court date.
* **Protective orders**: incident timeline, evidence, current safety plan, current order status.

**FR-3.4 Evidence and documents intake**

* Upload flow for key docs: prior orders, petitions, notices, parenting plans, financial statements, police reports (if applicable).
* Extract metadata (dates, court, case number) into structured fields.

**FR-3.5 Data validation + completeness scoring**

* Validate required fields per matter type.
* Display progress and ask only missing essentials.
* Compute “Intake Completeness Score” (e.g., 0–100) for staff.

---

## 4) Agentic orchestration (multi-step workflow behavior)

**FR-4.1 Plan-and-execute intake flow**

* Agent dynamically decides next question based on:

  * Matter type confidence
  * Missing required fields
  * Contradictions (e.g., “never married” but “date of marriage” provided)

**FR-4.2 Clarification and contradiction handling**

* Detect conflicting user statements.
* Ask a single clarifying question, then continue with corrected state.

**FR-4.3 Tool-use boundaries**

* Agent may use tools for: retrieval, deadline computation, conflict checks, scheduling, CRM/matter creation drafts.
* Agent must never finalize “legal conclusion”—only “preliminary qualification + attorney review”.

## 5) Agentic RAG knowledge retrieval requirements

**FR-5.1 Knowledge sources (RAG corpus)**
Must support retrieval from:

* Firm policies: what cases the firm accepts/declines, fee structures, geography served.
* Intake playbooks: required facts per matter type.
* Jurisdiction rules library (state-specific): filing requirements, residency rules, local court links.
* **Statute of limitations / deadlines library** (see next section).

**FR-5.2 Retrieval quality controls**

* Return citations (source doc + section) to staff-facing outputs.
* Confidence scoring and “insufficient authority” fallback.
* Versioning: track effective dates of rules; prefer most recent.

**FR-5.3 User-facing vs staff-facing content**

* User-facing: simplified procedural explanations.
* Staff-facing: exact rule text snippets, citations, rationale, and risk flags.

## 6) Deadline and SOL-like logic for family law (rules + computation)

Family law often has **deadlines** (service response windows, appeal deadlines, notice requirements, relocation notice, protective order hearing windows), even when “statute of limitations” isn’t the main driver. Build it broadly as “deadline engine”.

**FR-6.1 Jurisdiction resolution**

* Determine governing state (and sometimes county) based on:

  * Where filing would occur
  * Where children reside (custody jurisdiction)
* If ambiguous: capture both and mark “attorney review required”.

**FR-6.2 Deadline rules retrieval**

* Retrieve relevant deadlines based on matter type + jurisdiction + stage:

  * served with petition? upcoming hearing? existing order violated? relocation planned?
* Rules must be stored as structured logic when possible (not only text).

**FR-6.3 Date extraction**

* Extract key dates from chat + documents:

  * Date served, incident date, order date, last modification date, hearing date, move date.
* If missing: agent asks for the date explicitly.

**FR-6.4 Deadline calculation**

* Compute deadlines with:

  * business day vs calendar day handling (jurisdiction-specific)
  * configurable holiday calendar (MVP can ignore holidays but must mark as assumption)
* Output: computed due date, days remaining, severity level.

**FR-6.5 Near-deadline flagging**

* Threshold-based alerts (configurable, e.g., <30 days = “watch”, <14 = “urgent”, <7 = “critical”).
* Add “risk of prejudice” label when applicable.

**FR-6.6 “Assumption disclosure”**

* If the system computed using assumptions (unknown service method, unknown county), it must display:

  * assumptions used
  * what information would change the result

**FR-7.1 Qualification criteria configuration**

* Asset complexity thresholds (e.g., business valuation, high net worth)

**FR-7.2 Conflict pre-check integration**

* Search existing clients/opposing parties against firm CRM/case management.
* Match on name + phone/email + fuzzy matching.
* Output: “no match / possible match / match” with escalation to staff.

**FR-7.3 Urgency + priority scoring**

* Combine:

  * deadline proximity
  * safety risk

## 3) Recommended technology stack (MVP-friendly)

### Frontend

* **Chat widget**: React + TypeScript (embedded script)
* **UI**: component-driven prompts (cards, quick replies, date picker)
* **Streaming**: WebSocket or SSE for responsive chat

### Backend (recommended: AWS serverless MVP)

* **API Gateway** + **Lambda** (BFF + Orchestrator endpoints)
* **Step Functions** (optional) for workflow orchestration (staff queues, memo generation)
* **SQS** for async tasks (notifications, memo creation)
* **RDS Postgres** (or Aurora Serverless) for structured intake + search
* **S3** for transcript/memo artifacts
* **CloudWatch** for logs + metrics

(Equivalent alternatives: Azure Functions + Service Bus + Cosmos/SQL, or GCP Cloud Run + Pub/Sub + Cloud SQL.)

### LLM layer

* **Primary LLM**: OpenAI (chat + summarization)
* Use **structured outputs** (JSON schema) for:

  * matter type classification output
  * extracted fields
  * memo format
  * reason codes
* Guardrails:

  * “No legal advice” system policy
  * response templates + disclaimer injection
  * strict schema validation before persisting

### RAG layer

* **Vector DB**: Pinecone / Weaviate / pgvector (Postgres extension)
* **Indexing pipeline**:

  * chunk policy + deadline docs
  * embed and store with metadata (doc_id, version, section)
* Retrieval:

  * top-k + metadata filters (Florida-only, policy vs deadline library)

### Deadline engine

* Python or Node microservice/module:

  * date parsing (dateutil / chrono-node)
  * computation (20-day rule)
  * “near deadline” threshold config
  * outputs: computed deadline, rule_id, estimated=true

### Conflict check (MVP)

* Postgres tables: `clients`, `consults`, `matters`
* Name normalization:

  * lowercase, strip punctuation, split tokens
* Fuzzy matching:

  * trigram similarity (Postgres pg_trgm) or RapidFuzz
* Output:

  * `NO_MATCH` / `POSSIBLE_MATCH` + matched record IDs for staff only

### Scheduling handoff

* **Calendly/Acuity link** (MVP)
* Optional: create “pre-filled” scheduling page via query params
* Log scheduling click + completion webhook (if available)

### Notifications

* Email: SendGrid / AWS SES
* SMS (optional): Twilio
* Internal: Slack/Teams webhook for emergencies

### Security & compliance essentials (MVP)

* Encrypt at rest (RDS/S3), TLS in transit
* RBAC for staff console
* PII minimization in LLM prompts (send only needed fields)
* Audit log (append-only) for decisions and admin changes
* Data retention policy controls
