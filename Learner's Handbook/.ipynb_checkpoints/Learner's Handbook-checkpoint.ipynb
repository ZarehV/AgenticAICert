# Program Structure

## Introduction

Welcome to the 16 Week learning journey of JHU Agentic AI Program! This program is designed to immerse you in a structured, weekly learning rhythm that balances theory, practical skills, and real-world applications, all aimed at making you confident and proficient in Agentic AI. Over the next few weeks, you’ll progress from a recap of Python programming to large language models, advanced techniques in prompt engineering and retrieval-augmented generation, and core concepts of agentic AI, culminating in building and designing autonomous and advanced agentic AI systems for diverse business scenarios.

Let’s dive in, embrace the rhythm, and make this journey into the world of Agentic AI an exciting, transformative experience.

## Learning Outcomes

- Understand the Foundations of Agentic AI
- Develop Proficiency in Python for AI Applications
- Implement Prompt Engineering & RAG Techniques
- Design and Construct Agentic Systems
- Evaluate Ethical and Safety Considerations in AI Design
- Build Single-Agent and Multi-Agent Systems 
- Implement Advanced Learning Paradigms in AI Agents
- Explore Human-Agent Interaction Principles
- Critically Assess Limitations and Future Directions of Agentic AI
- Build Real-World Applications Using Agentic AI Frameworks

## Program Structure



## Learning Journey

The program consists of the following weeks:

1. Python Refresher for Agentic AI: This week focuses on setting up the Python environment and understanding core Python syntax, object-oriented programming concepts, and essential libraries for data science.

2. Introduction to Large Language Models: In this week, you will learn what large language models are, how they function, and the foundational concepts behind their architecture. Additionally, you will explore various prompt engineering techniques, including zero-shot prompting, one-shot prompting, and few-shot prompting.

3. Prompt Engineering Techniques & RAG: In this week, you will master various prompt engineering techniques necessary for the foundations of building and designing Agentic AI systems, including methods like Chain of Thought (CoT) and Retrieval-Augmented Generation (RAG).

4. Introduction to Agentic AI: In this week, you will learn what AI agents are and their core components. You will also explore how agentic AI differs from prior paradigms such as traditional Machine Learning, Deep Learning, and Generative AI. By understanding these distinctions, you'll gain insight into the evolution of AI systems and the unique capabilities that agentic AI brings to problem-solving.

5. Core Agent Concepts: This week teaches you to utilize the PEAS framework (Performance, Environment, Actuators, Sensors) to formally define agent tasks and environments, along with classifying agent architectures based on complexity and capabilities.

6. Ethics, Safety, Alignment & Responsible AI: In this week, you will evaluate the ethical implications of agentic AI, understand the AI alignment problem, and explore the core principles of responsible AI to ensure ethical development.

7. Planning and Reasoning Mechanisms:  This week covers the distinction between reasoning and planning, exploring advanced reasoning paradigms and the challenges of reliable agent evaluation, also focuses on building the Single-Agent Systems.

8. Multi-Agent Systems: In this week, you will define multi-agent systems and examine the unique challenges they present, including communication, coordination, and strategic interactions among agents.

9. Reinforcement learning in Agents: This week focuses on the fundamental concepts of reinforcement learning, differentiating between value-based and policy-based algorithms, and understanding when to use learning-based agents.

10. Interaction and Embodiment: In this week, you will explore human-agent interaction principles that foster trust and collaboration while analyzing the challenges of embodied AI and the simulation-to-real world gap.

11. The Future of Agentic AI: This week looks at the current limitations of agentic AI, discussing research challenges and the future landscape of agent technologies.

## Learning Instruments

This program offers various learning instruments that should be used together to achieve the most effective outcome

Recorded Videos: Recorded videos are the backbone of your course, providing structured and comprehensive coverage of essential concepts relevant to the respective weeks. Each week, you'll have access to recorded lectures tailored to the weekly curriculum, allowing you to learn at your own pace. To maximize your learning experience, immerse yourself in these recorded videos, understanding the concepts thoroughly. The knowledge gained here will be crucial for your success in upcoming Mentored Learning Sessions and assessments. 
 

Concept Capsule: These are designed to revisit fundamental topics and give you the first hints about their relevance to the main concept covered in that week. 
 

Test Your Understanding (TYUs): You will encounter embedded quizzes known as "Test Your Understanding" along with the recorded videos to encourage you to check your understanding of the concepts covered. Take these quizzes as you progress with the recorded videos to ensure you are following the material and reinforcing your learning. These quizzes are ungraded and are meant to check/test your understanding corresponding to the topics taught in the recorded videos.
 

Hands-on Handbooks: After completion of recorded videos, you will be provided with a handbook teaching them the tool/code/algorithm and implementing it using guided instructions explaining each step in detail. This will form the base before you go to the next section, i.e., MLS.
 

Practice Exercises: Engage with practice exercises to practice and reinforce your learning. Completing these helps you apply the concepts learned with respect to the particular week and test your understanding in a practical context. Make sure that you attempt all the practice exercises available before your Mentored Learning Sessions to reflect on your understanding corresponding to the week.
 

Mentored Learning Sessions (MLS): The live, instructor-led Mentored Learning Sessions (MLS) are a key component of this program. Participation in MLS is critical to fully grasp the concepts and prepare for assessments. Please note that attendance in these sessions is mandatory and carries significant weight in your overall grading. We encourage active engagement and preparation before each session by reviewing the relevant content. You will engage with your mentor to:
-Clarify doubts related to course content
-Collaborate on case studies
-Apply theoretical knowledge to practical problems
 

Graded Quizzes: These are mandatory weekly assessments designed to assess your understanding and track your learning progress. Complete these quizzes promptly after your Mentored Learning Sessions to evaluate your knowledge and receive grades, keeping you on track. Approach these quizzes with sincerity, attempt them in one go, and use your best understanding of the week's concepts. If your scores are lower than expected, use this feedback to identify areas for improvement and practice the concepts further to enhance your understanding.
 

Graded Projects / Assignments: Work on hands-on projects that simulate real-world scenarios. These projects provide practical experience and help you apply the knowledge learned from the learning content and sessions to solve complex problems. 
 

Discussion Forums: Join discussion forums to interact with peers and engage in collaborative learning. Participate in discussions, ask questions, and share insights with your peers to enhance your understanding and learn from others.
 

Additional Reading Material/ References: These materials enhance your understanding and keep you posted with additional content that can enhance your learning journey. You will be provided with additional reading materials/references wherever required throughout the learning journey.
